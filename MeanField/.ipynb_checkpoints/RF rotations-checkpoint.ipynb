```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import ode
from numpy.lib import scimath
%matplotlib inline
```

```python
def rf(t,y,*args):
    """rf ode"""
    c1 = y[0]
    c0 = y[1]
    cm1 = y[2]
    s = 1j*rabi/np.sqrt(2) 
    #now define equations
    f0 = s * c0 *np.cos(omega*t) np.exp(-1j*dm*t)
    f1 = s*cm1*np.exp(-1j*dp*t)+s*c1*np.exp(1j*dm*t)
    f2 = s*c0*np.exp(1j*dp*t)
    return [f0,f1,f2]   
```

```python
rabi = 10
dm = 1
dp = dm
def normalize_state(y):
    y = np.asarray(y)
    norm = np.abs(np.dot(np.conj(y),y))
    return y/norm
y0 = [np.complex(.2,0),np.complex(1,0),np.complex(.8,0)]
y0 = normalize_state(y0)
print(y0)
r = ode(rf).set_integrator('zvode')
r.set_initial_value(y0,0)
dt = 1e-3
tfinal = 2*np.pi / rabi+.1
t = np.linspace(0,tfinal, tfinal/dt+1) 
ans = np.zeros((len(t),3),dtype = complex)
step  = 0
while r.successful() and r.t < tfinal:
    ans[step] = np.asarray(r.integrate(r.t + dt))
    step += 1
```

```python
toplot = (np.conj(ans) * ans).real

plt.plot(t,toplot[:,0], label = r'$c_1$')
plt.plot(t,toplot[:,1], label = r'$c_0$')
plt.plot(t,toplot[:,2], label = r'$c_{-1}$')
plt.legend()
plt.axvline(2*np.pi/rabi,0,1,ls='--',c='black')
plt.axvline(np.pi/rabi,0,1,ls='--',c='black')
```
